```
# Librerias
import mglearn
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline 
```

```
mglearn.plots.plot_pca_illustration()
```

```
# Cargo los datos de una tabla de cancer y visualizo los distintos targets
import numpy as np
import pandas as pd
#
cancer=load_breast_cancer()
print("Numero de atributos de la tabla = "+str(cancer.feature_names.shape[0]))
#
pd.DataFrame(cancer.feature_names).T
```

```
# Pedimos que la disperción de puntos existentes en la tabla del PCA
# Encuentre las 2 componentes principales existentes que encierra la data
pca=PCA(n_components=2)
pca.fit(cancer.data)
pca.explained_variance_
```

```
# Ahora que transforme la tabla de 30 columnas a solamente una tabla de
# 2 columna que identifica los 2 componentes principales.
transformada=pca.transform(cancer.data)
#
print("Numero de atributos del dataset sin PCA = "+str(cancer.data.shape[1]))
print("Numero de atributos del dataset con PCA = "+str(transformada.shape[1]))
```

```
# Definimos X,Y
X = np.array(transformada)
Y = np.array(cancer.target)[:, np.newaxis]
# Graficamos
plt.scatter(transformada[Y[:,0] == 0,0],transformada[Y[:,0] == 0,1], color='red',  marker='^', alpha=0.5) 
plt.scatter(transformada[Y[:,0] == 1,0],transformada[Y[:,0] == 1,1], color='blue', marker='o', alpha=0.5) 
plt.grid(True)
plt.title("transformada")
plt.xlabel("PCA_1")
plt.ylabel("PCA_2")
plt.yscale('linear')
plt.xscale('linear')
plt.tight_layout()
plt.legend()
plt.show()
```
